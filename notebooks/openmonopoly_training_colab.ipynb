## 4. Inspect metrics


```python
import json
from pathlib import Path

metrics = json.loads(Path('outputs/colab_demo/training_metrics.json').read_text())
rollout = json.loads(Path('outputs/colab_demo/best_rollout.json').read_text())

print('Episodes per trained agent:', metrics['episodes_per_trained_agent'])
print('Average reward first 10:', metrics['avg_reward_first_10'])
print('Average reward last 10:', metrics['avg_reward_last_10'])
print('Best net worth:', metrics['best_net_worth'])
print('Showcase winner:', rollout['winner_name'])
print('Showcase turns:', rollout['turns_played'])
print('Termination:', rollout['termination_reason'])
```
